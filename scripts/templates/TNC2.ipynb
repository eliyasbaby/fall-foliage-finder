#### Background and VIC runs used

This notebook is for the flood frequency analysis project with The Nature Conservancy, progress report produced at the end of Dec, 2013. The VIC run used in this notebook is completed over the globe for 1948-1958, with 0.25 degree climate forcings and calibrated soil parameters from J. Sheffield, using VIC 4.0.6 in Full Energy mode. The first year of results were not used to account for the spin-up period. The calibration of the soil parameters was done to match the runoff fields at a grid cell level for selected basins, using VIC 4.0.5 in Full Energy mode. 

```python
# Import all needed modules
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import datetime
import numpy as np
import pandas as pd
from mpl_toolkits.basemap import Basemap, cm
from netCDF4 import Dataset, num2date, date2num, default_fillvals
```

#### Plots using the routed VIC simulated results in daily time series.

```python
# Routed, monthly streamflow directory for 1948-1958
#routed_dir = '/nfs/therapy/raid/tcchiao/TNC/vic.global.025.FE/10year_test/results/routed/'
routed_dir='/Users/nijssen/data/TNC/vic.global.025.FE/10year_test/results/routed/'
# Names of basins that we have data in
basins = ['colorado', 'columbia', 'mekong', 'mississippi', 'zambezi']
# Names of the streamflow observation stations used for each basin
stations = ['GLENC', 'Dalle', 'Pakse', 'Herm ', 'Senan']

# Monthly observed streamflow directory for 1949-1958
# Not used because simulated flow is very different from observed
#observed_dir = '/nfs/therapy/raid/tcchiao/TNC/vic.global.025.FE/10year_test/monthly.obs.flow/'
# Monthly time series is available for 4 of the 5 basins listed above (except Mekong)
# This is because we don't have data for Mekong for the time period simulated 
# However, there is mean monthly observed flow available for Mekong in the directory

# Load all the simulated flow into a dictionary 
simflow = {}
for i, (basin,station) in enumerate(zip(basins, stations)):
    # Get the right directory paths for each basin/station
    sim_file = routed_dir + basin + '/rout/output.txt/' + station +'.day'
    # Load the flows, unit = cfs
    # To get total flow in units of mm, use ".day_mm" as the final file extension
    simflow[basin] = np.loadtxt(sim_file, usecols = [3])

# Convert the dictionary into a Pandas DataFrame with the time stamps in the last file
# The time stamps only had to be loaded once because all simulations are of the same time period
   
# Load the time stamps
yy, mm, dd = np.loadtxt(sim_file, dtype = int, usecols = (0,1,2), unpack=True)
# Convert the string format dates into a list of datetime objects
simdates = []
for i in range(len(yy)):
    simdates.append(datetime.datetime(yy[i],mm[i],dd[i]))
# Do the conversion and remove the first two years of record to account for the spin-up period
simflow = pd.DataFrame(simflow, index = simdates).truncate(before = simdates[366])
cfstom3 = 0.3048**3
simflow *= cfstom3
simflow.describe()
```

### Flow duration curve and annual max flood recurrence interval curves (NOT normalized)

```python
# Plot the flow duration curve for these 5 basins
# List of line colors to use
colors = ['brown', 'r', 'orange', 'g', 'b']
basinnames = ['Colorado', 'Columbia', 'Mekong', 'Mississippi', 'Zambezi']

fig = plt.figure(figsize(12.5,4.5))
ax = fig.add_subplot(1,2,1)
# Loop through each basin
for i, basin in enumerate(basins):
    # Sort the daily flows
    flows = sort(simflow[basin].values)
    # Calculate the probability of exceedence based on the rank of each daily flow
    rank = np.arange(1, len(flows)+1, dtype = float)
    prob = 100-(100*(rank/len(flows)))
    # Plot
    ax.plot(prob, flows, color = colors[i], linewidth = 2.5, label = basinnames[i])
ax.set_ylabel('Daily Flow $(m^3)$', fontsize = 15.)
ax.set_xlabel('Percent Time of Exceedence', fontsize = 15.)
ax.set_yscale('log')
ax.set_ylim(1)
ax.tick_params(axis='both', which='major', labelsize=12.)
plt.tight_layout(True)


ax.legend(prop = {'size':14.}, bbox_to_anchor=(2.75, 0.75))

ax = fig.add_subplot(1,2,2)
for i, basin in enumerate(basins):
    # Sort the daily flows, find the max values for each year
    flows = np.sort(simflow[basin].groupby(lambda x: x.year).max().values)
    # Calculate the recurrence period based on Cunnane Recurrence interval
    n = len(flows)
    rank = (np.arange(len(flows), 0, -1, dtype = float))
    Tr = (n+0.2)/(rank-0.4)
    # Plot
    ax.scatter(Tr, flows, color = colors[i])
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlim(1,20)
ax.set_ylim(100,100000)
ax.xaxis.set_major_locator(ticker.FixedLocator([1,2,5,10,15]))
ax.xaxis.set_major_formatter(ticker.FixedFormatter(['1','2','5','10','15']))
ax.set_ylabel('Annual Flood Discharge $(m^3)$', fontsize = 15.)
ax.set_xlabel('Recurrence Interval (years)', fontsize = 15.)
ax.tick_params(axis='both', which='major', labelsize=12.)
plt.tight_layout(True)


fig.subplots_adjust(right = 0.75, wspace = 0.22)

#savefig('../figures/flow.duration.annual.flood.recurrence.png', dpi = 150, format = 'png')
```

### Flow duration curve and annual max flood recurrence interval curves (Normalized by basin area). 

```python
# Plot the flow duration curve for these 5 basins
# List of line colors to use
colors = ['brown', 'r', 'orange', 'g', 'b']
basinnames = ['Colorado', 'Columbia', 'Mekong', 'Mississippi', 'Zambezi']
# Basin area in km^2 
basinareas = [637000., 668000., 795000., 2981000., 1390000.] 
secspday = 24*3600

fig = plt.figure(figsize(12.5,4.5))
ax = fig.add_subplot(1,2,1)
# Loop through each basin
for i, basin in enumerate(basins):
    # Sort the daily flows, normalized by basin area
    flows = sort(simflow[basin].values)/basinareas[i]
    flows *= secspday/1000.
    # Calculate the probability of exceedence based on the rank of each daily flow
    rank = np.arange(1, len(flows)+1, dtype = float)
    prob = 100-(100*(rank/len(flows)))
    # Plot
    ax.plot(prob, flows, color = colors[i], linewidth = 2.5, label = basinnames[i])
ax.set_ylabel('Daily Flow $(mm/day)$', fontsize = 15.)
ax.set_xlabel('Percent Time of Exceedence', fontsize = 15.)
ax.tick_params(axis='both', which='major', labelsize=12.)
plt.tight_layout(True)
ax.legend(prop = {'size':15.}, bbox_to_anchor=(2.85, 0.75))

subplot(1,2,2)
for i, basin in enumerate(basins):
    # Sort the daily flows, find the max values for each year, normalize by basin area 
    flows = np.sort(simflow[basin].groupby(lambda x: x.year).max().values)/basinareas[i]
    flows *= secspday/1000.
    # Calculate the recurrence period based on Cunnane Recurrence interval
    n = len(flows)
    rank = (np.arange(len(flows), 0, -1, dtype = float))
    Tr = (n+0.2)/(rank-0.4)
    # Plot
    scatter(Tr, flows, color = colors[i])
ax = gca()
ax.set_xscale('log')
ylabel('Annual Flood Discharge $(mm/day)$', fontsize = 15.)
xlabel('Recurrence Interval (years)', fontsize = 15.)
tick_params(axis='both', which='major', labelsize=12.)
tight_layout(True)
xlim(1,20)
ylim(0,6)
ax.xaxis.set_major_locator(ticker.FixedLocator([1,2,5,10,15]))
ax.xaxis.set_major_formatter(ticker.FixedFormatter(['1','2','5','10','15']))

subplots_adjust(right = 0.75, wspace = 0.22)

#savefig('../figures/flow.duration.annual.flood.recurrence.png', dpi = 150, format = 'png')
```

### Maps of 10 year seasonal-average precipitation, evaporation, total runoff, and SWE. 

```python
# Set default font size for matplotlib plots
mpl.rcParams.update({'font.size': 16})
mpl.rc('xtick', labelsize=12)
mpl.rc('ytick', labelsize=12)
mpl.rc('legend', fontsize=12)
```

```python
#infile = '/nfs/therapy/raid/tcchiao/TNC/vic.global.025.FE/10year_test/results/netcdf/1949.1958.seasonal.mean.'
infile = '/Users/nijssen/data/TNC/vic.global.025.FE/10year_test/results/netcdf/1949.1958.seasonal.mean.'
ncvars = ['Precipitation', 'Evaporation', 'Total.Runoff']
cmap = cm.GMT_drywet
norms = [mpl.colors.Normalize(0, 10), mpl.colors.Normalize(0, 5), mpl.colors.Normalize(0, 5)]
area_thresh=25000
land_color='grey'
ocean_color='lightblue'
seasons = ['DJF', 'MAM', 'JJA', 'SON']


for i, ncvar in enumerate(ncvars):
    nc = Dataset(infile+ncvar+'.nc', 'r')
    lats = nc.variables['lat'][:]
    lons = nc.variables['lon'][:]
    var = nc.variables[ncvar][:]
    nc.close()

    fig = plt.figure(figsize=(18,7.5))
    
    for j,val in enumerate(seasons):
        ax = fig.add_subplot(2,2,j)
        m = Basemap(projection='cyl', llcrnrlat=lats.min(), llcrnrlon=lons.min(),
                urcrnrlat=lats.max(), urcrnrlon=lons.max(), resolution='i',
                area_thresh=area_thresh)
        xi, yi = m(lons, lats)
        xi, yi = np.meshgrid(xi, yi)
        m.pcolormesh(xi, yi, var[j], cmap=cmap, norm=norms[i])
        m.drawlsmask(land_color=land_color, ocean_color=ocean_color, lakes=True)
        m.drawcountries()
        m.drawcoastlines()
        ax.text(0.05, 0.05, val, horizontalalignment='left',
            verticalalignment='bottom', transform=ax.transAxes, fontsize=20)
    # Note that the white spaces between the subplots are also controlled by the aspect ratio of the projected basemaps
    # This means that matplotlib tries to conserve the aspect ratio of the projected basemaps and will create white spaces 
    # if necessary based on the figure size
    fig.subplots_adjust(wspace = 0.05, hspace = 0.05)
    
    fig = plt.figure(figsize=(10, 2))
    # using a linear colormap
    ax1 = fig.add_axes([0.05, 0.80, 0.9, 0.15])
    cbar = mpl.colorbar.ColorbarBase(ax1, cmap=cmap, norm=norms[i], orientation='horizontal', extend='max')
    cbar.set_label('{} (mm/day)'.format(ncvar), fontsize=16)
```

```python
#infile = '/nfs/therapy/raid/tcchiao/TNC/vic.global.025.FE/10year_test/results/netcdf/1949.1958.seasonal.mean.'
infile = '/Users/nijssen/data/TNC/vic.global.025.FE/10year_test/results/netcdf/1949.1958.seasonal.mean.'
ncvar = 'SWE'
cmap = mpl.cm.get_cmap('bone')
norm = mpl.colors.Normalize(0, 50)
area_thresh=25000
land_color=(135./255., 97./255., 42./255.)
ocean_color='lightblue'
seasons = ['DJF', 'MAM', 'JJA', 'SON']


nc = Dataset(infile+ncvar+'.nc', 'r')
lats = nc.variables['lat'][:]
lons = nc.variables['lon'][:]
var = nc.variables[ncvar][:]
var[var<=5] = np.ma.masked
nc.close()

fig = plt.figure(figsize=(18,7.5))
    
for i,val in enumerate(seasons):
    ax = fig.add_subplot(2,2,i)
    m = Basemap(projection='cyl', llcrnrlat=lats.min(), llcrnrlon=lons.min(),
            urcrnrlat=lats.max(), urcrnrlon=lons.max(), resolution='i',
            area_thresh=area_thresh)
    xi, yi = m(lons, lats)
    xi, yi = np.meshgrid(xi, yi)
    m.pcolormesh(xi, yi, var[i], cmap=cmap, norm=norm)
    m.drawlsmask(land_color=land_color, ocean_color=ocean_color, lakes=False)
    m.drawcountries()
    m.drawcoastlines()
    ax.text(0.05, 0.05, val, horizontalalignment='left',
        verticalalignment='bottom', transform=ax.transAxes, fontsize=20)
fig.subplots_adjust(wspace = 0.05, hspace = 0.05)
    
fig = plt.figure(figsize=(10, 2))
# using a linear colormap
ax1 = fig.add_axes([0.05, 0.80, 0.9, 0.15])
cbar = mpl.colorbar.ColorbarBase(ax1, cmap=cmap, norm=norm, orientation='horizontal', extend='max')
cbar.set_label('{} (mm)'.format('Snow water equivalent'), fontsize=16)
```
