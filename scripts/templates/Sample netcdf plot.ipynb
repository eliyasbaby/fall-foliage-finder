Sample global plot based on NetCDF file

```python
# import a bunch of modules that we are going to use anyway
import matplotlib as mpl
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap, cm
import numpy as np
from netCDF4 import Dataset, num2date, date2num
```

```python
# Set default font size for matplotlib plots
mpl.rcParams.update({'font.size': 16})
mpl.rc('xtick', labelsize=12)
mpl.rc('ytick', labelsize=12)
mpl.rc('legend', fontsize=12)
```

```python
infile = '/Users/nijssen/junk/1949.1958.seasonal.mean.evap.nc'
ncvar = 'Evaporation'
cmap = cm.GMT_drywet
norm = mpl.colors.Normalize(0, 5)
area_thresh=25000
land_color='grey'
ocean_color='lightblue'
seasons = ['DJF', 'MAM', 'JJA', 'SON']

nc = Dataset(infile, 'r')
lats = nc.variables['lat'][:]
lons = nc.variables['lon'][:]
var = nc.variables[ncvar][:]
nc.close()

fig = plt.figure(figsize=(20,12))
for i,val in enumerate(seasons):
    ax = fig.add_subplot(2,2,i)
    m = Basemap(projection='cyl', llcrnrlat=lats.min(), llcrnrlon=lons.min(),
            urcrnrlat=lats.max(), urcrnrlon=lons.max(), resolution='i',
            area_thresh=area_thresh)
    xi, yi = m(lons, lats)
    xi, yi = np.meshgrid(xi, yi)
    m.pcolormesh(xi, yi, var[i], cmap=cmap, norm=norm)
    m.drawlsmask(land_color=land_color, ocean_color=ocean_color, lakes=True)
    m.drawcountries()
    m.drawcoastlines()
    ax.text(0.05, 0.05, val, horizontalalignment='left',
        verticalalignment='bottom', transform=ax.transAxes, fontsize=20)
```

```python
fig = plt.figure(figsize=(10, 3))
# using a linear colormap
ax1 = fig.add_axes([0.05, 0.80, 0.9, 0.15])
cbar = mpl.colorbar.ColorbarBase(ax1, cmap=cmap, norm=norm, orientation='horizontal', extend='max')
cbar.set_label('{} (mm/day)'.format(ncvar), fontsize=16)
```
